```python
# prompt: Реализуйте метод наименьших квадратов для линейной регрессии: даны точки (1,2), (2,3), (3,5), (4,4), (5,6). Найдите коэффициенты прямой  y=ax+b  используя NumPy и визуализируйте результат.

import matplotlib.pyplot as plt
import numpy as np
# Data points
x_points = np.array([1, 2, 3, 4, 5])
y_points = np.array([2, 3, 5, 4, 6])

# Design matrix X
# For a linear regression y = ax + b, the design matrix has two columns:
# the first column is a column of ones (for the intercept b),
# and the second column is the x values (for the slope a).
X = np.vstack([np.ones(len(x_points)), x_points]).T

# Calculate coefficients (a and b) using the normal equation: coefficients = (X^T * X)^(-1) * X^T * y
# X^T is the transpose of X
# np.linalg.inv() calculates the inverse of a matrix
# np.dot() calculates matrix multiplication
coefficients = np.linalg.inv(X.T @ X) @ X.T @ y_points

# The coefficients are in the order [b, a]
b_coeff, a_coeff = coefficients

print(f"Коэффициенты прямой y = ax + b:")
print(f"a = {a_coeff}")
print(f"b = {b_coeff}")

# Generate points for the fitted line
x_fit = np.linspace(x_points.min(), x_points.max(), 100)
y_fit = a_coeff * x_fit + b_coeff

# Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(x_points, y_points, label='Исходные точки', color='blue')
plt.plot(x_fit, y_fit, color='red', label=f'Линейная регрессия: y = {a_coeff:.2f}x + {b_coeff:.2f}')
plt.title('Метод наименьших квадратов для линейной регрессии')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Решите ОДУ первого порядка  dydx=x2−2y  с начальным условием  y(0)=1  на интервале [0, 5] используя SciPy.

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

# Определяем функцию правой части ОДУ
def ode_func(x, y):
  return x**2 - 2*y

# Определяем интервал интегрирования
x_span = [0, 5]

# Определяем начальное условие y(0) = 1
y0 = [1]

# Решаем ОДУ
solution = solve_ivp(ode_func, x_span, y0, dense_output=True)

# Создаем массив точек x для построения графика решения
x_plot = np.linspace(x_span[0], x_span[1], 100)

# Вычисляем значения y на основе интерполированного решения
y_plot = solution.sol(x_plot)[0]

# Строим график решения
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='Решение ОДУ')
plt.scatter(x_span[0], y0[0], color='red', label='Начальное условие y(0)=1') # Отмечаем начальную точку
plt.title('Решение ОДУ dydx = x^2 - 2y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Выводим несколько значений решения для проверки
print("Несколько значений решения:")
for x_val in np.linspace(0, 5, 11):
    y_val = solution.sol(x_val)[0]
    print(f"y({x_val:.1f}) = {y_val:.4f}")
```

```python
# prompt: Постройте 3D-поверхность функции  z=x2−y2+xy  для  x,y∈[−3,3]  используя Matplotlib, найдите и отметьте точки экстремума.

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
  return x**2 - y**2 + x*y

# Create grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the 3D surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Find the critical points by finding the gradient and setting it to zero
# The gradient of f(x, y) is (df/dx, df/dy)
# df/dx = 2x + y
# df/dy = -2y + x
# We need to solve the system of equations:
# 2x + y = 0
# x - 2y = 0
# From the first equation, y = -2x. Substitute into the second equation:
# x - 2(-2x) = 0
# x + 4x = 0
# 5x = 0
# x = 0
# Substituting x = 0 back into y = -2x gives y = 0.
# So the only critical point is at (0, 0).

# Check the second derivative test for the critical point (0, 0)
# d^2f/dx^2 = 2
# d^2f/dy^2 = -2
# d^2f/dxdy = 1
# The Hessian matrix is [[2, 1], [1, -2]]
# The determinant of the Hessian is (2)(-2) - (1)(1) = -4 - 1 = -5
# Since the determinant is negative, the critical point (0, 0) is a saddle point, not a local extremum.
# Therefore, there are no local extrema.

# Although there are no local extrema, the code below shows how to mark a point on the surface if needed
# You can change the point (x_crit, y_crit) if you find other critical points for a different function.
# x_crit = 0
# y_crit = 0
# z_crit = f(x_crit, y_crit)
# ax.scatter(x_crit, y_crit, z_crit, color='red', s=100, label='Критическая точка')

ax.set_title('3D-поверхность функции z = x² - y² + xy')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# ax.legend()
plt.show()
```

```python
# prompt: Реализуйте алгоритм k-средних для набора точек на плоскости, используя SciPy для кластеризации и Matplotlib для визуализации результатов с разными цветами кластеров.

import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.vq import kmeans, vq

# Генерируем случайный набор точек на плоскости
np.random.seed(42) # для воспроизводимости результатов
points = np.random.rand(100, 2) * 10 # 100 точек с координатами от 0 до 10

# Задаем количество кластеров
k = 3

# Применяем алгоритм k-средних
centroids, distortion = kmeans(points, k)

# Определяем принадлежность каждой точки к кластеру
cluster_indices, _ = vq(points, centroids)

# Визуализируем результаты
plt.figure(figsize=(8, 6))

# Отображаем точки, раскрашивая их в зависимости от принадлежности к кластеру
for i in range(k):
    plt.plot(points[cluster_indices == i, 0], points[cluster_indices == i, 1], 'o', label=f'Кластер {i+1}')

# Отображаем центроиды кластеров
plt.plot(centroids[:, 0], centroids[:, 1], 'kx', markersize=15, mew=3, label='Центроиды')

plt.title('Результаты кластеризации K-средних')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
