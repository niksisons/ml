Категоризируйте возраст пассажиров "Титаника" на три группы:

1. **Категория 1**: Молодые пассажиры (до 30 лет)
   - Если возраст меньше 30, функция возвращает 1.

2. **Категория 2**: Пассажиры среднего возраста (от 30 до 54 лет)
   - Если возраст от 30 до 54 включительно, функция возвращает 2.

3. **Категория 3**: Пожилые пассажиры (55 лет и старше)
   - Для всех остальных случаев (то есть 55 лет и старше) функция возвращает 3.

```python
def ages(age):
    match age:
        case age if age < 30:
            return 1
        case age if 30 <= age <= 54:
            return 2
        case _:
            return 3

df['AgeCy'] = df['Age'].apply(ages)
print(df[['Age', 'AgeC']].head(30))
```

**4. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?**
- 22.7% среди молодых и 40.6% среди старых
- 40.6% среди молодых и 22.7% среди старых
- 35.3% среди молодых и 27.4% среди старых
- 27.4% среди молодых и 35.3% среди старых

```python
young_group = df[df['Age'] < 30]
old_group = df[df['Age'] > 60]

young = young_group['Survived'].mean()
old = old_group['Survived'].mean()

print(f"Доля выживших среди людей младше 30 лет: {young:.2f}")
print(f"Доля выживших среди людей старше 60 лет: {old:.2f}")

if young > old:
    print("Правда")
else:
    print("Неправда")
```

**5. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?**

- 21.1% среди мужчин и 46.2% среди женщин
